###     1. Реализовать программу выполняющую операции сложения, умножения, деления и вычитания над любыми числами. Необходимо учесть ситуацию деления на ноль, с помощью конструкции if избежать выполнения такой операции и сообщить пользователю об этом.
## Чек лист самопроверки:
###     1. Программа предлагает ввести два любых числа.
###     2. Программа производит выбранную математическую операцию над введенными числами и возвращает результат.
###     3. Если пользователь хочет поделить число на ноль, программа должна сообщить об этом, не выполняя запрашиваемую операцию.

```python
def arithmetic_operations():
    print("Приветствую!")
    while True:
        try:
            a, b = int(input("Введите значение a: ")), int(input("Введите значение b: "))
            operation = input("Введите необходимую операцию + - / *: ")
            if operation == '+':
                print(f'a + b = {a + b}')
            elif operation == '-':
                print(f'a - b = {a - b}')
            elif operation == '/':
                if b == 0:
                    print("На нуль делить нельзя!")
                else:
                    print(f'a // b = {a // b}')
            elif operation == '*':
                print(f'a * b = {a * b}')
            else:
                print("Операция невозможна")
        except Exception as exc:
            print(f"Возникла ошибка: {exc}")
        repeat = input("Повторить? [y/n]: ")
        if repeat.lower() == 'n': break
    print("До скорого!")
    
arithmetic_operations()
```

## Задание:
###    1. Реализовать программу выполняющую сравнение двух последовательностей поэлементно, при чем такую, что цикл while вложен в цикл for, и когда совпадающие символы найдены, программа должна об этом сообщить.Также нужно указать индексы совпадающих элементов для каждой последовательности. Нельзя использовать функцию enumerate.
## Чек лист самопроверки:
###     1. Программа предлагает ввести две любых строки (последовательности).
###     2. Программа возвращает результат – совпадающие символы двух последовательностей, или то, что таких символов не найдено.
###     3. Программа также выводит индексы совпадающих элементов для каждой последовательности.

```python
def cmp_sequences():
    print("Приветствую!")
    while True:
        try:
            s1, s2 = input("Введите последовательность 1: "),  input("Введите последовательность 2: ")

            idx1 = 0
            idx2 = 0
            count = 0
            for с in s1:
                while idx2 < len(s2) and idx1 < len(s1):
                    if s1[idx1] == s2[idx2]:
                        print(f"Найдено совпадение символа '{s2[idx2]}' по индексу {idx2}")
                        count += 1
                    idx1 += 1
                    idx2 += 1
            print(f"Количество совпадений: {count} " if count != 0 else "Совпадений не найдено")
        except Exception as exc:
            print(f"Возникла ошибка: {exc}")
        repeat = input("Повторить? [y/n]: ")
        if repeat.lower() == 'n': break
    print("До скорого!")
    
cmp_sequences()
```

Реализовать программу принимающую от пользователя любую последовательность и с помощью функций map и lambda меняющую элементы, таким образом,что если элемент число,тогда производится операция умножения на два, а если элемент имеет строковый тип, тогда производится размножение элемента в два раза. Результат работы возвращать в виде списка.
## Чек лист самопроверки:
###     1.Программа предлагает ввести последовательность.
###     2.Программа возвращает результат работы в виде списка, при этом, если числа должны быть увеличены в два раза, то строковые элементы размножены в два раза.

```python
def double_mult():
    print("Приветствую!")
    while True:
        str_test = input("Введите последовательность для типа строка: ")
        list_test = list(int(c) if c.isdigit() else c for c in input("Введите последовательность для типа лист: "))
        tuple_test = tuple(int(c) if c.isdigit() else c for c in input("Введите последовательность для типа кортеж: "))

        str_result = list(map(lambda i: i * 2, str_test))
        list_result = list(map(lambda i: i * 2, list_test))
        tuple__result = tuple(map(lambda i: i * 2, tuple_test))
        print(f"Результат со строкой: {str_result} \nРезультат со списком: {list_result} \n"
              f"Результат с кортежем: {tuple__result} \n")
        repeat = input("Повторить? [y/n]: ")
        if repeat.lower() == 'n': break
    print("До скорого!")
    
double_mult()
```

```python
# Отображение кол-ва пустых значений
def show_empty_values(df):
    count = 0
    for i in df:
        percnt_empty = (len(df[i]) - df[i].count())/len(df[i])
        if percnt_empty != 0:
            print(f"{i} - содержит {round(percnt_empty*100,2)}% пропусков.")
            count += 1
    return count
```

```python
# Удаление пустых значений
df2 = df.dropna()
```
